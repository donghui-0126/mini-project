```python
data = {
    'index': list(df.index),
    '1min_ago_change': df['1min_ago_change']
}

df = pd.DataFrame(data)

# 'index' ì—´ì„ datetime í˜•ì‹ìœ¼ë¡œ ë³€í™˜
df['index'] = pd.to_datetime(df['index'])

# 'index' ì—´ì„ ë‚ ì§œë¡œ ì„¤ì •í•˜ì—¬ ë°ì´í„°ë¥¼ ê·¸ë£¹í™”í•˜ê³ , ê° ë‚ ì§œì— ëŒ€í•œ ë³€ë™ì„± ê³„ì‚°
df.set_index('index', inplace=True)  # 'index' ì—´ì„ ì¸ë±ìŠ¤ë¡œ ì„¤ì •

# ë‚ ì§œë³„ë¡œ ë³€ë™ì„± ê³„ì‚°
daily_volatility = df.resample('D')['1min_ago_change'].std()

print(daily_volatility)
```
