```python
import pandas as pd
```

```python
df = pd.read_csv("train_data_2023.csv", index_col=0)
```

```python
df.head(15)
```

```python
df[abs(df["1min_ago_change"])>0.0008].index.to_list()
```

```python
data = {
    'index': list(df.index),
    '1min_ago_change': df['1min_ago_change']
}

df = pd.DataFrame(data)

# 'index' 열을 datetime 형식으로 변환
df['index'] = pd.to_datetime(df['index'])

# 'index' 열을 날짜로 설정하여 데이터를 그룹화하고, 각 날짜에 대한 변동성 계산
df.set_index('index', inplace=True)  # 'index' 열을 인덱스로 설정

# 날짜별로 변동성 계산
daily_volatility = df.resample('D')['1min_ago_change'].std()

print(daily_volatility)
```

```python
H_volatility = df.resample('H')['1min_ago_change'].std()

print(H_volatility)
```

```python
W_volatility = df.resample('W')['1min_ago_change'].std()

print(W_volatility)
```

```python
M_volatility = df.resample('M')['1min_ago_change'].std()

print(M_volatility)
```

```python
Y_volatility = df.resample('Y')['1min_ago_change'].std()

print(Y_volatility)
```
